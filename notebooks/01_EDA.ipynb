```python
📌 Cell 1 — Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

📌 Cell 2 — Load Dataset
from src.preprocessing.data_loader import load_data

df = load_data("../data/raw/bank_transactions.csv")

df.head()

Output

transaction_id	transaction_amount	merchant	label
1001	550	Amazon	0
1002	12000	Crypto	1

📌 Cell 3 — Shape of Dataset

print("Rows and Columns:")
print(df.shape)

Example Output

(10000, 15)

📌 Cell 4 — Dataset Information
df.info()

Output

<class 'pandas.core.frame.DataFrame'>

10000 entries

15 columns

📌 Cell 5 — Statistical Summary
df.describe()


📌 Cell 6 — Column Names
df.columns


📌 Cell 7 — Check Missing Values
df.isnull().sum()


📌 Cell 8 — Fill Missing Values
df["transaction_amount"] = df["transaction_amount"].fillna(
    df["transaction_amount"].median()
)

df["merchant"] = df["merchant"].fillna("Unknown")


📌 Cell 9 — Check Duplicate Records
print("Duplicate Rows:")


print(df.duplicated().sum())
📌 Cell 10 — Remove Duplicates
df = df.drop_duplicates()


📌 Cell 11 — Class Distribution
df["label"].value_counts()

Example

0    9500

1     500


📌 Cell 12 — Fraud Percentage
fraud_percentage = df["label"].value_counts(normalize=True) * 100

fraud_percentage

Output

Legitimate : 95%

Fraud : 5%



📌 Cell 13 — Histogram
plt.figure(figsize=(8,5))

plt.hist(df["transaction_amount"], bins=30)

plt.title("Transaction Amount Distribution")

plt.xlabel("Transaction Amount")

plt.ylabel("Frequency")

plt.show()



📌 Cell 14 — Boxplot
plt.figure(figsize=(6,4))

plt.boxplot(df["transaction_amount"])

plt.title("Outlier Detection")





plt.show()
📌 Cell 15 — Correlation Matrix
numeric_df = df.select_dtypes(include="number")

corr = numeric_df.corr()

corr



📌 Cell 16 — Correlation Heatmap
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

plt.imshow(corr, cmap="coolwarm")

plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)

plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap")

plt.show()



📌 Cell 17 — Save Clean Dataset
df.to_csv("../data/processed/clean_transactions.csv", index=False)

print("Dataset Saved Successfully!")
📊 Final Notebook Structure
01_EDA.ipynb

Cell 1 → Import Libraries
Cell 2 → Load Dataset
Cell 3 → Shape
Cell 4 → Info
Cell 5 → Describe
Cell 6 → Columns
Cell 7 → Missing Values
Cell 8 → Fill Missing Values
Cell 9 → Duplicate Check
Cell 10 → Remove Duplicates
Cell 11 → Class Distribution
Cell 12 → Fraud Percentage
Cell 13 → Histogram
Cell 14 → Boxplot
Cell 15 → Correlation Matrix
Cell 16 → Heatmap
Cell 17 → Save Clean Dataset
```
